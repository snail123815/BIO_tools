```python
idtopacc = {}
for i, r in anno.iterrows():
    if r[0] not in idtopacc:
        idtopacc[r[0]] = [r[1],r[3]]
    else:
        acccurr = r[1]
        ecurr = r[3]
        accstor = idtopacc[r[0]][0]
        estor = idtopacc[r[0]][1]
        baseacccurr = acccurr.split(':')[0]
        baseaccstor = accstor.split(':')[0]
        if ecurr == estor:
            if baseacccurr == baseaccstor:
                if not (":" in acccurr and ":" in accstor):
                    accchos = baseacccurr # accession chosen
                else:
                    accchos = accstor # first appearance
            else: # different base accession, choose the none subfamily one, else choose first appearance
                if (int(":" in acccurr) + int(":" in accstor)) % 2 == 0:
                    accchos = accstor
                elif ":" not in acccurr:
                    accchos = acccurr
                else:
                    accchos = accstor
            echos = ecurr # e value chosen
        else:
            if ecurr > estor:
                accchos = acccurr
                echos = ecurr
            else:
                accchos = accstor
                echos = estor
        idtopacc[r[0]] = [accchos, echos]
pd.DataFrame(idtopacc).T
```
