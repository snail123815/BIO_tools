```python
allGene = {} # Every gene into this dict, using SCO number as key
for annotations in listOutput:
    geneNames = annotations[0].strip().split(': ')[1].strip().split(', ')
    while '' in geneNames: # Remove some junk
        geneNames.remove('')
    
    geneIDs = [name for name in geneNames if name.startswith('SCO')] # This is a list because there are some genes 'SCO0022', 'SCO0023', 'SCO0024' are annotated at the same position
    for geneID in geneIDs: 
        geneNames.remove(geneID)
        
    oldIDs = [name for name in geneNames if name.startswith('SC') or name.startswith('3SC') or name.startswith('2SC') or name.startswith('St')]
    for oldID in oldIDs:
        geneNames.remove(oldID)
    
    refs = [] # grab reference list for each gene, link for every reference is not extracted as googling is much easier
    for strReference in re.findall(r'<li>(.*?)</li>', annotations[2]):
        reference = re.findall(r'<span[^>]*>(.*?)</span>', strReference)
        reference.insert(1 , re.findall(r'</span>(.*?)<span[^>]*>', strReference)[1].strip())
        refs.append(reference)
    if len(refs) != 0:
        for refi, ref in enumerate(refs):
            refs[refi][3] = ref[3].replace('<em>', '').replace('</em>', '')    
            
    if len(geneIDs) > 1:
        for geneID in geneIDs:
            allGene[geneID] = [geneIDs, geneNames, oldIDs, annotations[1].strip(), refs]
    else:
        allGene[geneIDs[0]] = [geneIDs, geneNames, oldIDs, annotations[1].strip(), refs]

allGene['SCO3034']
```
